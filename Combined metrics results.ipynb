```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define your models' evaluation metrics, including H-CapsNet, Zero-Shot, and N-Shot
metrics_data = {
    "Model": ["BERT", "BART", "RoBERTa", "H-CapsNet", "Zero-Shot", "N-Shot"],
    "Accuracy": [0.3684, 0.3994, 0.3003, 0.1522, 0.4733, 0.8769],
    "Precision": [0.7400, 0.7334, 0.7302, 0.4316, 0.4967, 0.8779],
    "Recall": [0.6841, 0.7045, 0.6716, 0.7611, 0.4733, 0.8769],
    "F1-Score": [0.7110, 0.7187, 0.6997, 0.5134, 0.4209, 0.876989],
    "Test Error Rate": [0.6316, 0.6006, 0.6997, 0.8478, 0.5267, 0.1231]
}

# Convert the dictionary into a DataFrame
metrics_df = pd.DataFrame(metrics_data)

# Define the labels and metrics for the grouped bar chart
models = metrics_df['Model']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'Test Error Rate']

# Create a grouped bar chart
bar_width = 0.15
index = np.arange(len(models))

# Set figure size
plt.figure(figsize=(12, 8))

# Plotting each metric as a separate set of bars
for i, metric in enumerate(metrics):
    plt.bar(index + i * bar_width, metrics_df[metric], bar_width, label=metric)

# Adding labels and title
plt.xlabel('Models', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.title('Comparison of Model Performance Metrics', fontsize=14)
plt.xticks(index + bar_width * 2, models)  # Center the x-ticks
plt.legend(title="Metrics")

# Display the grouped bar chart
plt.show()
```
