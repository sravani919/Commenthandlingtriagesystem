```python
import torch
import pandas as pd
from transformers import BertForSequenceClassification, BertTokenizer
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import f1_score
import numpy as np

# Load the saved model
model = BertForSequenceClassification.from_pretrained('./model2_save')

# Load the tokenizer
tokenizer = BertTokenizer.from_pretrained('./model2_save')

# Set the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Step 1: Load your labeled dataset
df = pd.read_csv("labeled_comments_cleaned.csv")

# Define the category names for each level
level_categories = {
    'level_0': ['INFORMATION EXCHANGE', 'MODIFICATION', 'SOCIAL COMMUNICATION', 'OTHER'],
    'level_1': ['REQUESTED', 'EXECUTION', 'DISCUSSION', 'PROVIDED', 'FEEDBACK', 'ACKNOWLEDGMENT'],
    'level_2': ['REQUESTING_CONFIRMATION', 'PROMISE', 'CONTENT', 'REFERENCE', 'CONTEXT', 'THREAD', 'DONE', 'FORMAT', 'ASKING_DETAILS'],
    'level_3': ['POTENTIAL_CHANGE', 'NOT_POTENTIAL_CHANGE', 'EXPLICIT', 'NOT_EXPLICIT'],
    'level_4': ['CHANGE', 'ADD', 'DELETE']
}

# Step 2: Tokenize the test texts
test_texts = df['comment_sentence_text'].tolist()
encodings = tokenizer(test_texts, padding=True, truncation=True, return_tensors='pt')

# Step 3: Create evaluation function for category-wise F1 score for multiple levels
def evaluate_f1_for_all_levels(model, data_loader, device, df, level_categories):
    results = []

    for level, categories in level_categories.items():
        # Prepare one-hot encoded true labels for the current level
        true_labels = pd.get_dummies(df[level]).values  # One-hot encoding of the level
        labels_tensor = torch.tensor(true_labels)

        # Create a new DataLoader for this level's labels
        test_dataset = TensorDataset(encodings['input_ids'], encodings['attention_mask'], labels_tensor)
        test_loader = DataLoader(test_dataset, batch_size=16)

        model.eval()  # Set the model to evaluation mode
        predictions, true_labels_all = [], []

        with torch.no_grad():
            for batch in test_loader:
                input_ids = batch[0].to(device)
                attention_mask = batch[1].to(device)
                labels = batch[2].to(device)

                # Get model outputs
                outputs = model(input_ids, attention_mask=attention_mask).logits
                sigmoid_outputs = torch.sigmoid(outputs)
                pred_bools = (sigmoid_outputs > 0.5).cpu().numpy()

                predictions.append(pred_bools)
                true_labels_all.append(labels.cpu().numpy())

        # Flatten the predictions and true labels lists
        predictions = np.vstack(predictions)
        true_labels_all = np.vstack(true_labels_all)

        # Calculate F1 score for each category in the current level
        category_f1_scores = {}
        for i, category in enumerate(categories):  # Use provided category names
            f1 = f1_score(true_labels_all[:, i], predictions[:, i], average='weighted')
            category_f1_scores[category] = f1

        # Add results for this level to the list
        results.append(pd.DataFrame(list(category_f1_scores.items()), columns=['Category', f'F1_Score_{level}']))

    return results

# Step 4: Run the evaluation and get F1 scores for each level
results = evaluate_f1_for_all_levels(model, encodings, device, df, level_categories)

# Step 5: Combine the F1 scores from all levels into one DataFrame
f1_df_combined = pd.concat(results, axis=0, ignore_index=True)

# Step 6: Display the combined F1 score distribution
print(f1_df_combined)

# Optionally, save the F1 score distribution to a CSV
f1_df_combined.to_csv("all_levels_f1_scores_by_category.csv", index=False)
```

```python
import torch
import pandas as pd
from transformers import BertForSequenceClassification, BertTokenizer
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import f1_score
import numpy as np

# Load the saved model
model = BertForSequenceClassification.from_pretrained('./model2_save')

# Load the tokenizer
tokenizer = BertTokenizer.from_pretrained('./model2_save')

# Set the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Step 1: Load your labeled dataset
df = pd.read_csv("labeled_comments_cleaned.csv")

# Define the category names for each level
level_categories = {
    'level_0': ['INFORMATION EXCHANGE', 'MODIFICATION', 'SOCIAL COMMUNICATION', 'OTHER'],
    'level_1': ['REQUESTED', 'EXECUTION', 'DISCUSSION', 'PROVIDED', 'FEEDBACK', 'ACKNOWLEDGMENT'],
    'level_2': ['REQUESTING_CONFIRMATION', 'PROMISE', 'CONTENT', 'REFERENCE', 'CONTEXT', 'THREAD', 'DONE', 'FORMAT', 'ASKING_DETAILS'],
    'level_3': ['POTENTIAL_CHANGE', 'NOT_POTENTIAL_CHANGE', 'EXPLICIT', 'NOT_EXPLICIT'],
    'level_4': ['CHANGE', 'ADD', 'DELETE']
}

# Step 2: Tokenize the test texts
test_texts = df['comment_sentence_text'].tolist()
encodings = tokenizer(test_texts, padding=True, truncation=True, return_tensors='pt')

# Step 3: Create evaluation function for category-wise F1 score for multiple levels
def evaluate_f1_for_all_levels(model, data_loader, device, df, level_categories):
    all_results = pd.DataFrame()  # Initialize an empty DataFrame to store all results

    for level, categories in level_categories.items():
        # Prepare one-hot encoded true labels for the current level
        true_labels = pd.get_dummies(df[level]).values  # One-hot encoding of the level
        labels_tensor = torch.tensor(true_labels)

        # Create a new DataLoader for this level's labels
        test_dataset = TensorDataset(encodings['input_ids'], encodings['attention_mask'], labels_tensor)
        test_loader = DataLoader(test_dataset, batch_size=16)

        model.eval()  # Set the model to evaluation mode
        predictions, true_labels_all = [], []

        with torch.no_grad():
            for batch in test_loader:
                input_ids = batch[0].to(device)
                attention_mask = batch[1].to(device)
                labels = batch[2].to(device)

                # Get model outputs
                outputs = model(input_ids, attention_mask=attention_mask).logits
                sigmoid_outputs = torch.sigmoid(outputs)
                pred_bools = (sigmoid_outputs > 0.5).cpu().numpy()

                predictions.append(pred_bools)
                true_labels_all.append(labels.cpu().numpy())

        # Flatten the predictions and true labels lists
        predictions = np.vstack(predictions)
        true_labels_all = np.vstack(true_labels_all)

        # Create a dictionary to store F1 scores for each category in the current level
        category_f1_scores = {}
        for i, category in enumerate(categories):  # Use provided category names
            f1 = f1_score(true_labels_all[:, i], predictions[:, i], average='weighted')
            category_f1_scores[category] = f1

        # Convert F1 scores to DataFrame
        level_results = pd.DataFrame(list(category_f1_scores.items()), columns=['Category', f'F1_Score_{level}'])

        # Merge results into the combined DataFrame
        if all_results.empty:
            all_results = level_results
        else:
            all_results = pd.merge(all_results, level_results, on='Category', how='outer')

    return all_results

# Step 4: Run the evaluation and get F1 scores for each level
f1_df_combined = evaluate_f1_for_all_levels(model, encodings, device, df, level_categories)

# Step 5: Display the combined F1 score distribution
print(f1_df_combined)

# Optionally, save the F1 score distribution to a CSV
f1_df_combined.to_csv("all_levels_f1_scores_by_category.csv", index=False)
```
